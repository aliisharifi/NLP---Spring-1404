```python
import requests
from bs4 import BeautifulSoup
import os
```

```python

# Define the website URL
url = "https://fa.wikipedia.org/wiki/%D8%A7%D8%B5%D9%81%D9%87%D8%A7%D9%86"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
# Find all paragraphs with the specific style attribute
paragraphs = soup.find_all(["p", "h2", "h3", "h4"])

#for item in paragraphs:
#    print(f"  {item}")

wiki_text = "\n".join(p.text.strip() for p in paragraphs)
print("๐ฐ Extracted Wiki Content:")
print(wiki_text)
```

```python
#file_path = r"C:\D\Learn\Uni\MS-Sharif\_Term2\NLP\HW\HW1\NLP---Spring-1404\Crawl" 

folder = "../Isfahan"
filename = "ุงุตููุงู.txt"

os.makedirs(folder, exist_ok=True)

file_path = os.path.join(folder, filename)

with open(file_path, "w", encoding="utf-8") as file:
    file.write(wiki_text)
    print("File Saved!")
```
