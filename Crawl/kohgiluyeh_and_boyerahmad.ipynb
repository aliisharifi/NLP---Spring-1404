## geographical_features

### رودخانه‌ ها  

```python
#رود خانه ها
URL = "https://ecotourism.blog.ir/1393/10/21-3"

response = requests.get(URL)
soup = BeautifulSoup(response.text, "html.parser")

paragraphs = soup.find_all("p")
text += "\n" + "رود خانه ها"

for p in paragraphs:
    strong_tags = p.find_all("strong")
    for tag in strong_tags:
        content = tag.get_text(strip=True)
        if content:
            if content.endswith(":"):
                content = content[:-1]
            text += "\n" + content
            print(content)
```

### کوه ها

```python
#کوه ها
URL = "https://ecotourism.blog.ir/1393/10/21-1"

response = requests.get(URL)
soup = BeautifulSoup(response.text, "html.parser")

paragraphs = soup.find_all("p")
text += "\n" + "کوه ها"

for p in paragraphs:
    strong_tags = p.find_all("strong")
    for tag in strong_tags:
        content = tag.get_text(strip=True)
        if content and content.startswith("ک"):
            if content.endswith(":"):
                content = content[:-1]
            text += "\n" + content
            print(content)
```

```python
URL = "https://www.flytoday.ir/blog/kohgiluyeh-and-boyer-ahmad-attractions/"

response = requests.get(URL)
soup = BeautifulSoup(response.text, "html.parser")

h3s = soup.find_all("h3", class_="wp-block-heading")
paragraphs = soup.find_all("p")
img_divs = soup.find_all("div", class_="wp-block-image")

one_para_indices = {17, 28, 33, 34}

para_index = 5  
img_index = 2  

for idx, h3 in enumerate(h3s):
    title = h3.get_text(strip=True)
    print(title)

    num_paras = 1 if idx in one_para_indices else 2

    for _ in range(num_paras):
        if para_index < len(paragraphs):
            content = paragraphs[para_index].get_text(strip=True)
            print(content)
            para_index += 1

    if img_index < len(img_divs):
        img_tag = img_divs[img_index].find("img")
        if img_tag and img_tag.has_attr("src"):
            src = img_tag['src']
            print(src)
        img_index += 1
```
